```python
import numpy as np
import matplotlib.pyplot as plt
 
xp = np.linspace(-1, 1, 100)
yp = np.array([np.sqrt(x0 + 1) - np.cos(x0) for x0 in xp])
plt.plot(xp, yp, '--')
 
x = np.linspace(-1, 1, 5)
y = np.array([np.sqrt(x0 + 1) - np.cos(x0) for x0 in x])
plt.plot(x, y, 'o', markersize=5)
 
poly3 = np.poly1d(np.polyfit(x, y, 3))
plt.plot(xp, poly3(xp), '-')
 
leg = np.polynomial.legendre.legfit(x,y,3)
print(leg)
plt.plot(xp, np.polynomial.legendre.legval(xp, leg), '-')
 
plt.show()
```
